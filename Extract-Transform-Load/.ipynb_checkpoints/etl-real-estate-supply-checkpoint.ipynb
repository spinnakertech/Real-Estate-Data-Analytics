```python
# Pull National data into new dataframe for the desired columns
national_df = raw_data_national_df[re_sales_columns]
national_df
```

```python
# Pull State and County data into new dataframe for the desired columns
state_county_df = raw_data_state_county_df[re_sales_columns]
state_county_df
```

```python
# Combine the National dataframe and the State and County dataframe
combined_df = national_df.append(state_county_df,ignore_index=True)
combined_df
```

```python
# Rename columns to prepare to load into RDBMS
real_estate_df = combined_df.rename(columns={"Region": "region",
                                             "Month of Period End": "month_year",
                                            "Region Type": "region_type",
                                            "State Code": "state_code",
                                            "State": "state_name",
                                            "Homes Sold": "homes_sold",
                                            "Homes Sold Yoy": "homes_sold_yoy",
                                            "Inventory": "homes_inventory",
                                            "Inventory Yoy": "homes_inventory_yoy",
                                            "Median Dom": "median_dom",
                                            "Median Dom Yoy": "median_dom_yoy",
                                            "Median Sale Price": "median_sale_price",
                                            "Median Sale Price Yoy": "median_sale_price_yoy",
                                            "New Listings": "new_listings",
                                            "New Listings Yoy": "new_listings_yoy"})
real_estate_df
```

```python
# Data clean up
# Split the 'Month of Period End' field into 'Month' and 'Year'
real_estate_df[['month','year']] = real_estate_df.month_year.str.split(" ",expand=True)

# Convert the median sales price to a number by removing the '$' and ','; and converting the 'K' to '000'
real_estate_df['median_sale_price'] = real_estate_df.median_sale_price.str.replace("$","")
real_estate_df['median_sale_price'] = real_estate_df.median_sale_price.str.replace(",","")
real_estate_df['median_sale_price'] = real_estate_df.median_sale_price.str.replace("K","000")
real_estate_df
```

```python
# Add a new month_code by referencing the month column for sorting for charting and plotting
# Create a dictionary of month_code keys & values
month_code_dict = {'January': '01',
                   'February': '02',
                   'March': '03',
                   'April': '04',
                   'May': '05',
                   'June': '06',
                   'July': '07',
                   'August': '08',
                   'September': '09',
                   'October': '10',
                   'November': '11',
                   'December': '12'
                  }

# Create a new column, month_code and set it's value to the month column
real_estate_df['month_code'] = real_estate_df['month']

# Remap the values of the new month_code field using the dictionary
real_estate_df['month_code'] = real_estate_df['month_code'].map(month_code_dict)

# Also add a month_year_code field to fully numeric for sorting
real_estate_df['month_year_code'] = real_estate_df['year'] + real_estate_df['month_code']

real_estate_df
```
